#Train the system using Wine quality dataset (winequality-red), determine the accuracy using Naive Bayes

Import Libraries

```python
import pandas as pd
import numpy as np
```

Load and read the dataset

```python
df=pd.read_csv("/content/winequality-red.csv", sep=';')

print("The first 5 values are: ")
print(df.head())
```

#EDA

```python
print("\nMissing Values: ")
print(df.isnull().sum())

print("\nInformation of wine dataset: ")
print(df.info())

print("\nDuplicate values: ")
print(df.duplicated().sum())

print("\nSize of the dataset: ")
print(df.size)
```

Assign values of X variable (independent variable) and y variables (dependent variable)

```python
X = df.drop("quality", axis=1)
y = df["quality"]
```

Train-test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Gaussian Model is used when the values are continous (numerical values)

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train, y_train)
```

```python
print("Accuracy:(of train) ", model.score(X_train, y_train))
print("test accuracy: ", model.score(X_test, y_test))
```
