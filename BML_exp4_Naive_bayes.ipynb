Assign values of X variable (independent variable) and y variable (dependent variable)

```python
X=df[['Pregnancies','Glucose','BloodPressure','Insulin','BMI','Age']]
y=df['Outcome']
```

Train-Test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

Model creation and model fitting

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train, y_train)
```

```python
#accuracy of the model for gaussian
print("Accuracy:(of train) ", model.score(X_train, y_train))
print("test accuracy: ", model.score(X_test, y_test))
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train, y_train)
```

```python
#accuracy of the model for multinomial
print("Accuracy:(of train) ", model.score(X_train, y_train))
print("test accuracy: ", model.score(X_test, y_test))
```

```python
from sklearn.naive_bayes import BernoulliNB
model=BernoulliNB()
model.fit(X_train, y_train)
```

```python
#accuracy of the model for bernoulli
print("Accuracy:(of train) ", model.score(X_train, y_train))
print("test accuracy: ", model.score(X_test, y_test))
```
